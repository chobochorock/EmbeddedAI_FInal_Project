```python
import yaml
import os

# ==========================================
# [설정] 경로를 본인 환경에 맞게 수정하세요
DATASET_DIR = os.path.abspath("../yolo_dataset")  # 아까 만든 데이터셋 폴더 절대 경로
CLASSES_FILE = os.path.join(DATASET_DIR, "classes.txt")
OUTPUT_YAML = "handwriting.yaml"
# ==========================================

def main():
    # 1. 클래스 이름 읽기
    with open(CLASSES_FILE, "r", encoding="utf-8") as f:
        class_names = [line.strip() for line in f.readlines()]
    
    print(f"총 {len(class_names)}개의 클래스를 로드했습니다.")

    # 2. YAML 데이터 구조 생성
    data = {
        'path': DATASET_DIR,          # 데이터셋 루트 경로
        'train': 'train/images',      # 학습 이미지 경로
        'val': 'train/images',        # 검증 이미지 (따로 없으면 train 사용)
        'nc': len(class_names),       # 클래스 개수 (4803)
        'names': class_names          # 클래스 이름 리스트
    }

    # 3. 파일 저장
    with open(OUTPUT_YAML, 'w', encoding="utf-8") as f:
        yaml.dump(data, f, allow_unicode=True)

    print(f"✅ 설정 파일 생성 완료: {OUTPUT_YAML}")
    print("이 파일을 yolov5 폴더 안으로 옮기거나 경로를 지정해서 학습하세요.")

if __name__ == "__main__":
    main()
```
