```python
import torch
from transformers import AutoTokenizer, AutoConfig, MobileBertForSequenceClassification
```

```python
# load model
model_name = r'cambridgeltl/sst_mobilebert-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name)
config = AutoConfig.from_pretrained(model_name)
model = MobileBertForSequenceClassification.from_pretrained(model_name, config=config).cuda()
model.eval()
'''
    labels: 
        0 -- negative
        1 -- neutral
        2 -- positive
'''
```

```python
# prepare exemplar sentences
batch_sentences = [
    "a very well-made , funny and entertaining picture .",
    "this movie is really boring, and not made well.",

]

# prepare input
inputs = tokenizer(batch_sentences, max_length=128, truncation=True, padding=True, return_tensors='pt')
input_ids, attention_mask = inputs.input_ids, inputs.attention_mask

# make predictions
outputs = model(input_ids=input_ids.cuda(), attention_mask=attention_mask.cuda())
predictions = torch.argmax(outputs.logits, dim = -1).cpu().numpy()
```

```python
p2l = {2:'positive', 1:'neutral', 0:'negative'}
print([p2l[x] for x in predictions])
```
